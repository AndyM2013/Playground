# Feature Transformation, NLP, and Classification Modeling on Hotel Rating

# Abstract

## The Incentive

The purpose of this post is that, in the prior attempt, I tried to build up a prediction model on reviewer's score for each hotel. It turns out the predictive power is not satisfactory, with $R^2$ (Coefficient of Determination) lingers around 0.4 to 0.5. Thus in this task, I'm ganna try a different path, in which I'll first discretize the reviewer's score into three ordinal categories - low, medium, high and then build up a classification model. See if it reaches higher predictive power under simplified classification method. It should be easier for the model to classify the reviewer's score into three levels instead of requiring to come up with precise predicted score.

## Information of The Dataset

The original dataset collected from Booking.com and is shared on [Kaggle](https://www.kaggle.com/jiashenliu/515k-hotel-reviews-data-in-europe) platform. This dataset contains 500 thousands reviews for hotels and collectes 14 features, including numerical socres and text reviews.

## Quick Summary of Model Performance

After comparing the performance of models among MLP, Xgboost, and Logistic, my final choice is **Logistic Model**. It turns out that Logistic Model not only outperforms than other, though only by a samll margin, but also has rather excellent efficiency on training. Hyperparameter fine tuning is also implemented, but the result shows no significant performance difference betweeen default configuration and optimized parameters setting. 

**The final model reaches 0.62 in accuracy and 0.61 in F1 score.** If we look deeper, the share of each class are 35%, 29%, 36%, which compared to the precision from the predicted result for each class 68%, 47%, 65%, we can say that **Logistic Model doubles the precision** generally.

**<mark>1. Preprocess Data</mark>**

Shown from the distribution, the features are highly skewed. It's better to use normal distributed features when feeding as inputs in the model. Thus, most numerical features will have **log transformation** and also **be scaled to have the same unit**, in preparation for modeling later on.

From models above, the best performed model is chosen as the final one, and then model's hyperparameters will be fine tuned via grid search, so that to further increase its accuracy. The final model will be tested on testing dataset and compare to the default configured model.

```python
from math import ceil
from scipy.stats import norm
from scipy import stats

def plot_normal(df, sub_col):
    """ Plot QQ-plot alongside histogram."""
    length = ceil(len(sub_col)/2)
    
    fig, axes = plt.subplots(length, 4, figsize=(20,17))
    axes = axes.ravel()
    
    index = 0
    for col in sub_col:
        (mu, sigma) = norm.fit(df[col])
        (a,b) = stats.probplot(df[col], sparams=(mu, sigma)) # QQ plot
        axes[index].plot(*a, color='#014d64', linewidth=5)
        (x_min, x_max) = axes[index].get_xlim()
        x_dots = np.arange(x_min, x_max)
        y_dots = b[0]*x_dots + b[1]
        axes[index].plot(x_dots, y_dots, color='#7c260b')
        axes[index].set_title('QQ Plot', fontsize='large')
        axes[index].set_xlabel('Theoretical Quantiles')
        axes[index].set_ylabel('Ordered Values')
        
        index += 1
        sns.distplot(df[col], ax=axes[index], kde=True, color='#c7e1e8', 
                     kde_kws={'color':'#014d64','label':'KDE'},
                     fit=norm, fit_kws={'color':'#7c260b','label':'NORMAL'}) # distribution plot
        axes[index].legend(loc='best')
        axes[index].set_title('Histogram', fontsize='large')
        index += 1
```

# Feature Engineering - Discretization

## Target Variable

Focused on target variable - reviewer_score, it will be divided into three groups. The bin interval are as follows,

<ol style='font-size:103%;line-height:25px'>
    <li>**Low: values<= 8.0.**</li>
    <li>**Medium: 8.0 < values <= 9.5.**</li>
    <li>**High: 9.5 < values <= 10.0.**</li>
</ol>

After discretization, target variable is split into three groups of each share is **35%, 29%, 36%** respectively.

```python
# create customized bin width
def bin_cut(df, col, bin_range):
    """Specify customized bin range for discretization."""
    target_bin = pd.cut(df[col], 
        bin_range, duplicates='drop') # drop off bins with the same index, incur less bin number

    print('Share of Each Bin:')
    bin_share = target_bin.groupby(target_bin).agg({'size': lambda x: x.size,
                                                    'share': lambda x: x.size/len(target_bin)})
    display(bin_share)

    map_class = {}
    for i, key in enumerate(sorted(target_bin.unique())):
        map_class[key] = i
    print('Bin and Class Label Correspondence:')
    display(map_class)

    target_bin_2 = target_bin.replace(map_class)
    target_bin = pd.concat([target_bin, target_bin_2], axis=1)
    target_bin.columns = ['{}_bin'.format(col), '{}_class'.format(col)]

    return target_bin
```

```python
discrete_target = bin_cut(df=clean_data, col='reviewer_score', bin_range=[0,8,9.5,10])
```

In discretization for features, I chose equal shares as the binning criterion, each bin has roughly the same amount of records.

```python
def bin_qcut(df, col, bins=10):
    """Discretize values based on euqal records in each bin."""
    print('Number of unique value of {}: {}\n'.format(col, len(df[col].unique())))
    
    fig, ax = plt.subplots(figsize=(15,5))
    sns.set_style(style='whitegrid')
    feature = df[col]
    _ = feature.plot.hist(ax=ax, bins=15, colors=['#cec3c6'])
    mean_line = ax.axvline(x=feature.mean(), linewidth=2, color='black', label='mean {:.2f}'.format(
        feature.mean()))
    median_line = ax.axvline(x=feature.median(), linewidth=2, color='red', label='median {:.2f}'.format(
        feature.median()))
    _ = ax.legend(handles=[mean_line, median_line])
    _ = ax.set_title('Histogram of {} for Bins = 15'.format(col), fontsize='x-large')
    
    plt.show()
    
    target_bin = pd.qcut(df[col], bins, duplicates='drop')
    bin_default = bins
    bin_final = len(target_bin.unique())
    bin_diff = bin_default - bin_final
    print('\n{} bins out of {} bins are dropped off.(some bins are dropped off because of overlapping bins)'.format(bin_diff, bin_default))
    
    print('Share of Each Bin:')
    bin_share = target_bin.groupby(target_bin).agg({'size': lambda x: x.size,
                                                    'share': lambda x: x.size/len(target_bin)})
    display(bin_share)
    
    map_class = {}
    for i, key in enumerate(sorted(target_bin.unique())):
        map_class[key] = i
    
    target_bin_2 = target_bin.replace(map_class)
    target_bin = pd.concat([target_bin, target_bin_2], axis=1)
    target_bin.columns = ['{}_bin'.format(col), '{}_class'.format(col)]

    return target_bin
```

## Lemmatization and Tokenization

Before tokenizing text, I use **lemmatizer** to further extract the root of each words, so that leave out the derivations of words in different forms. Then the stop words are neglected and to specify word selection threshold by `max_df` and `min_df`. Bi-gram is used in tokenization, so that we get to take into account the different meaning of phrases, such as `very good` and `not good`. First **2500** tokens are selected to be model input afterwards.

```python
# Create a class for lemmatizer.
class LemmaTokenizer(object):
    
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc)]
```

```python
# Only pick up the first 2500 words as word features.
n_features = 2500

tfidf_vectorizer = TfidfVectorizer(tokenizer=LemmaTokenizer(),
                                   max_df=0.8, min_df=10,
                                   max_features=n_features,
                                   stop_words='english',
                                   ngram_range=(1,2)) # select one and bi-grams
```

```python
# Fit and transform data.
tfidf_text = tfidf_vectorizer.fit_transform(text['combined_review'])
```

## MLP Model - Training

```python
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
```

```python
# Set up MLP model.
class MLP(nn.Module):
    def __init__(self, input_size, output_size, fc1_units=128, fc2_units=64):
        super().__init__()
        self.fc1 = nn.Linear(input_size, fc1_units)
        self.fc2 = nn.Linear(fc1_units, fc2_units)
        self.fc3 = nn.Linear(fc2_units, output_size)
    
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.softmax(self.fc3(x), dim=1)
        return x
```

```python
from math import floor

def get_batch(df, target, batch_size=32):
    """Helper function to get batch of data for training in Pytorch.
    
    @params
        df: scipy csr matrix
        target: pandas series
    
    """
    n_batch = floor(df.shape[0]/batch_size)
    
    complete_df = df[:n_batch*batch_size,:]
    y_target = target[:n_batch*batch_size]

    for i in range(0, n_batch*batch_size, batch_size):
        batch_feature = complete_df[i:i+batch_size, :]
        batch_y_target = y_target[i:i+batch_size]
        
        yield batch_feature, batch_y_target
```

```python
# Hyper-parameters.
input_size = fit_feature.shape[1]
output_size = 3
learning_rate = 0.02
batch_size = 32

epochs = 5
step = 0
print_every = 2500
```

```python
mlp_model = MLP(input_size, output_size)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(mlp_model.parameters(), lr=learning_rate)
```

```python
# train
for e in range(epochs):
    running_loss = 0
    generator = get_batch(df=x_train, target=y_train, batch_size=batch_size)
    for feature, y_true in generator:
        step += 1
        optimizer.zero_grad()
        
        feature = torch.tensor(np.array(feature.toarray()), dtype=torch.float)
        y_true = torch.tensor(np.array(y_true), dtype=torch.long)
        
        output = mlp_model(feature)
        loss = criterion(output, y_true)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
        if step % print_every == 0:
            print('Epoch: {}/{}...'.format(e+1, epochs),
                 'Loss: {:.4f}'.format(running_loss/print_every))
            
            running_loss = 0
```

## Fine Tuned Model - Training

```python
def heatmap(values, ylabel, xlabel, yticklabels, xticklabels, cmap=None,
            vmin=None, vmax=None, ax=None, fmt="%0.2f"):
    """Plot grid-search score results.
    
    @params
        values: list of mean_test_score being reshaped to square
        ylabel: name of first param (string)
        xlabel: name of second param (string)
        yticklabels: values of first param
        xticklabels: values of second param
    
    Ref: https://github.com/amueller/introduction_to_ml_with_python/blob/master/mglearn/tools.py
    """
    plt.figure(figsize=(10,10))
    if ax is None:
        ax = plt.gca()
    # plot the mean cross-validation scores
    img = ax.pcolor(values, cmap=cmap, vmin=vmin, vmax=vmax)
    img.update_scalarmappable()
    ax.set_xlabel(xlabel, fontsize='x-large')
    ax.set_ylabel(ylabel, fontsize='x-large')
    ax.set_xticks(np.arange(len(xticklabels)) + .5)
    ax.set_yticks(np.arange(len(yticklabels)) + .5)
    ax.set_xticklabels(xticklabels, fontsize='large')
    ax.set_yticklabels(yticklabels, fontsize='large')
    ax.set_aspect(1)

    for p, color, value in zip(img.get_paths(), img.get_facecolors(),
                               img.get_array()):
        x, y = p.vertices[:-2, :].mean(0)
        if np.mean(color[:3]) > 0.5:
            c = 'k'
        else:
            c = 'w'
        ax.text(x, y, fmt % value, color=c, ha="center", va="center")
    return img
```

```python
from sklearn.metrics import make_scorer, f1_score
from sklearn.model_selection import GridSearchCV

def fitGridModel(model, params, train_x, train_y):
    '''Define a function to implement GridSearch all in one.'''
    reg_model = model
    best_params = {}
    
    scorer = make_scorer(f1_score, greater_is_better=True, average='weighted') # multi-class averaging
    grid = GridSearchCV(reg_model, param_grid=params, scoring=scorer, cv=3, verbose=2)
    grid.fit(train_x, train_y)
    score = grid.best_score_
    print("the best f1 score of all model parameters' combination on model: {:.4f}".format(score))
    best_params.update(grid.best_params_)
    print("the parameter setting of optimized model:", grid.best_estimator_)
    
    return score, best_params, grid
```

```python
# Set up hyperparameters.
params = {'C':[0.01, 0.1, 0.5, 1, 5, 10],
          'penalty':['l1','l2']}
```

```python
# Fine-tune logistic model.
final_clf_score, final_clf_params, final_clf_model = fitGridModel(
             model=LogisticRegression(
                 random_state=1,
                 multi_class='multinomial',
                 solver='saga'),
             params=params,
             train_x=x_train,
             train_y=y_train)
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Plot heatmap of grid-search score.
scores = final_clf_model.cv_results_['mean_test_score'].reshape(6,2) # adjust shape according the dimension of params
param_C = params['C']
param_penalty = params['penalty']


# Set up customized color map.
# Ref: https://medium.com/@Vincentropy/a-practical-intro-to-colors-in-python-496737f23568
# Ref: http://colorbrewer2.org/#type=sequential&scheme=Purples&n=7
colors = ['#756bb1','#bcbddc','#efedf5']
linear_map = LinearSegmentedColormap.from_list(name='my_color', colors=colors)

printmd('**Score Matrix over All Params:**')
_ = heatmap(values=scores, 
            ylabel='C', 
            xlabel='penalty', 
            yticklabels=param_C,
            xticklabels=param_penalty,
            cmap=linear_map)
```
